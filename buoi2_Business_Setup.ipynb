## 🐍 Phần 2 — Cài đặt Virtual Environment & Dependencies

### 🤔 Virtual Environment là gì? Tại sao cần?

Tưởng tượng bạn có 2 dự án:
- Dự án A cần `pandas 1.5`
- Dự án B cần `pandas 2.2`

Nếu cài chung vào Python global → hai phiên bản xung đột. **Virtual environment (venv)** giải quyết bằng cách tạo "hộp Python riêng biệt" cho mỗi project.

```
Máy tính của bạn
├── Python (global)
├── venv-project-A/    ← pandas 1.5, numpy 1.23...
└── venv-project-B/    ← pandas 2.2, numpy 1.26...  ← dự án này dùng cái này
```

---

### ⚡ Step 3 — Tạo và kích hoạt Virtual Environment

**macOS / Linux:**
```bash
# Tạo virtual environment trong folder "venv"
python3 -m venv venv

# Kích hoạt virtual environment
source venv/bin/activate

# Dấu hiệu thành công: prompt terminal sẽ đổi thành (venv) $
# Ví dụ: (venv) user@mac cohort-retention-analytics $
```

**Windows (PowerShell):**
```powershell
# Tạo virtual environment
python -m venv venv

# Kích hoạt virtual environment
.\venv\Scripts\Activate.ps1

# Nếu bị lỗi "execution policy", chạy lệnh này trước:
Set-ExecutionPolicy -ExecutionPolicy RemoteSigned -Scope CurrentUser

# Dấu hiệu thành công: prompt đổi thành (venv) PS>
```

## 📄 Phần 3 — Viết `requirements.txt` Chuẩn Portfolio

### 🤔 requirements.txt là gì?

`requirements.txt` là file liệt kê **tất cả thư viện Python** mà project cần, kèm theo **phiên bản tối thiểu**. Bất kỳ ai nhận project của bạn chỉ cần chạy một lệnh là có đủ môi trường để reproduce kết quả.

```
Bạn chia sẻ project               Người khác nhận project
─────────────────────              ─────────────────────────
cohort-retention-analytics/   →   pip install -r requirements.txt
├── requirements.txt               → Tự động cài đúng phiên bản
├── notebooks/...                  → Chạy notebook không lỗi ✅
└── data/...
```

---

### ⚡ Step 6 — Tạo requirements.txt

**Cách 1 — Tạo thủ công (khuyến nghị cho học viên):**

**Cách 2 — Tự động sinh từ venv đang active (dùng sau khi đã cài đủ):**
```bash
pip freeze > requirements.txt
```

> ⚠️ **Lưu ý về Cách 2:** `pip freeze` sẽ ghi ra tất cả thư viện kèm phiên bản chính xác (ví dụ `pandas==2.2.3`). Phù hợp để "đóng băng" môi trường, nhưng dễ gây lỗi khi người khác dùng hệ điều hành khác. Với portfolio cá nhân, Cách 1 linh hoạt hơn.

### 🔍 Giải thích ký hiệu phiên bản

| Ký hiệu | Nghĩa | Ví dụ |
|---------|-------|-------|
| `>=2.2` | Phiên bản 2.2 trở lên đều được | `pandas>=2.2` |
| `==2.2.3` | Phải đúng phiên bản này | `pandas==2.2.3` |
| `>=2.2,<3.0` | Từ 2.2 đến dưới 3.0 | `pandas>=2.2,<3.0` |

> 💡 **Best practice:** Dùng `>=` thay vì `==` để không bị "pinned" vào một phiên bản quá cũ, trừ khi project có dependency conflict cụ thể.

## 📋 Phần 4 — Business Brief: Tại sao cần viết trước khi code?

### 🤔 Business Brief là gì?

**Business Brief** là tài liệu 1–2 trang viết **trước khi bắt đầu phân tích**, trả lời 3 câu hỏi:

1. **Context:** Tình huống kinh doanh là gì? Ai giao việc này, tại sao?
2. **Questions:** Cần trả lời cụ thể những câu hỏi gì?
3. **Decision:** Kết quả phân tích sẽ dùng để đưa ra quyết định gì?

---

### 💡 Tại sao phải viết Business Brief TRƯỚC KHI code?

Hãy so sánh 2 cách làm:

```
❌ Cách sai (hay gặp trong portfolio sinh viên):
   Mở Jupyter → viết code ngay → "EDA này, chart kia"
   → Cuối cùng: đống code không kể câu chuyện gì
   → Không trả lời được câu hỏi cụ thể nào của stakeholder

✅ Cách đúng (chuẩn Data Analyst thực tế):
   Viết Business Brief → Xác định metric → Lên outline notebook
   → Mỗi section code đều có mục đích rõ ràng
   → Kết quả phân tích trả lời đúng câu hỏi business
```

> 🎯 **Nguyên tắc:** Nếu bạn không thể giải thích một dòng code đang trả lời câu hỏi business nào, dòng code đó không nên có trong notebook.

---

### 🌍 Trong thực tế, Business Brief trông như thế nào?

Ở các công ty, document này có thể gọi là:
- *"Analysis brief"* / *"Data request"* (startup nhỏ)
- *"PRD - Product Requirements Document"* (product team)
- *"Analytics ticket"* (data team có Jira/Notion)
- *"Stakeholder requirement document"* (enterprise)

Dù tên khác nhau, nội dung cốt lõi đều giống nhau: **ai cần gì, để làm gì, khi nào cần**.

Trong dự án này, bạn sẽ tạo file `reports/business_brief.md` — một Markdown document đơn giản nhưng chuyên nghiệp.

## 📐 Phần 6 — Metric Dictionary: Định nghĩa KPI trước khi tính

### 🤔 Metric Dictionary là gì? Tại sao cần?

Đây là vấn đề **CỰC KỲ PHỔ BIẾN** trong thực tế:

```
Marketing Lead hỏi:  "Revenue tháng 3 là bao nhiêu?"
Analyst A trả lời:   "850 triệu"   ← tính cả đơn cancelled
Analyst B trả lời:   "790 triệu"   ← chỉ tính đơn completed
Analyst C trả lời:   "810 triệu"   ← tính net sau hoàn tiền

→ Ba con số khác nhau, ai đúng? Không ai sai — họ chỉ dùng định nghĩa khác nhau!
```

**Metric Dictionary** giải quyết bằng cách **đặt tên + định nghĩa chính xác mỗi metric** từ đầu, để toàn team dùng chung một con số.

---

### 📊 Metric Dictionary của dự án này

| Metric | Tên hiển thị | Công thức | Grain | Filter | Ghi chú |
|--------|-------------|-----------|-------|--------|---------|
| **Revenue** | Doanh thu | `SUM(net_revenue)` | Month / Channel / Cohort | `status == 'completed'` | Không tính đơn cancelled, returned |
| **Orders** | Số đơn hàng | `COUNT DISTINCT(order_id)` | Month / Channel | `status == 'completed'` | Không dùng row count nếu có order_items |
| **Active Customers** | Khách hàng active | `COUNT DISTINCT(customer_id)` | Month / Cohort | `status == 'completed'` | Dùng làm tử số tính retention |
| **AOV** | Giá trị đơn hàng TB | `Revenue / Orders` | Month / Channel | `orders > 0` | Average Order Value |
| **Cohort Size** | Kích thước cohort | `COUNT DISTINCT(customer_id)` tại `cohort_index = 0` | Cohort month | `status == 'completed'` | Mẫu số để tính retention rate |
| **Retention Rate** | Tỷ lệ giữ chân | `Active Customers / Cohort Size` | Cohort month + Index | — | Luôn là 100% tại index 0 |

---

### 🔍 Giải thích chi tiết từng khái niệm

**📌 `grain` là gì?**
> Grain = độ chi tiết nhỏ nhất của metric. Ví dụ: Revenue `grain = Month/Channel` có nghĩa là mỗi dòng trong bảng là revenue của **1 tháng + 1 channel** cụ thể. Grain nhỏ hơn → bảng nhiều dòng hơn, phân tích sâu hơn.

**📌 `filter` là gì?**
> Filter = điều kiện lọc dữ liệu trước khi tính. `status == 'completed'` nghĩa là bỏ qua tất cả đơn hàng bị hủy hoặc hoàn trả. Không note filter → người sau tính lại ra số khác → mất tin tưởng.

**📌 Cohort Index vs Cohort Month:**
```
Cohort Jan-2026 có 150 khách hàng:
  Cohort Month = "2026-01"      ← tháng khách gia nhập (không đổi)
  Cohort Index = 0, 1, 2, 3... ← số tháng tính từ ngày gia nhập

  Index 0: Jan/2026 → 150 khách active → 150/150 = 100%
  Index 1: Feb/2026 →  63 khách active →  63/150 = 42%
  Index 2: Mar/2026 →  41 khách active →  41/150 = 27%
  Index 3: Apr/2026 →  28 khách active →  28/150 = 19%
```

### 💻 Metric Dictionary trong code (preview Buổi 5)

Khi đến Buổi 5, bạn sẽ thấy mỗi metric trong dictionary ở trên map trực tiếp sang code Pandas:

```python
# Revenue — SUM(net_revenue), grain=month, filter=completed
monthly_revenue = (
    df[df['status'] == 'completed']   # ← filter
    .groupby(df['order_date'].dt.to_period('M'))  # ← grain=month
    ['net_revenue'].sum()              # ← SUM
    .reset_index()
)

# Active Customers — COUNT DISTINCT(customer_id), grain=month, filter=completed
monthly_customers = (
    df[df['status'] == 'completed']
    .groupby(df['order_date'].dt.to_period('M'))
    ['customer_id'].nunique()          # ← nunique() = COUNT DISTINCT
    .reset_index()
)

# AOV — Revenue / Orders, grain=month
monthly_kpi['aov'] = monthly_kpi['revenue'] / monthly_kpi['orders']
```

> 💡 Bạn chưa cần chạy được code này ngay — đây là minh họa để thấy **Metric Dictionary → Code** là một quá trình trực tiếp, không phải đoán mò.
