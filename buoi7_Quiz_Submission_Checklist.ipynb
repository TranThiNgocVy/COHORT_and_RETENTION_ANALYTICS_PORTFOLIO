**Câu 3.** Lệnh nào tạo thư mục nếu chưa tồn tại mà **không báo lỗi** khi thư mục đã có sẵn?

- A. `os.mkdir("folder")`
- B. `Path("folder").mkdir(parents=True, exist_ok=True)`
- C. `pd.mkdir("folder")`

**💡 Đáp án: B** — `exist_ok=True` giúp lệnh không báo lỗi nếu folder đã tồn tại. `parents=True` tự tạo cả thư mục cha nếu cần.

**Câu 4.** `parse_dates=["order_date"]` trong `pd.read_csv()` có tác dụng gì?

- A. Lọc chỉ lấy những dòng có ngày hợp lệ
- B. Chuyển cột `order_date` từ string thành kiểu ngày (datetime)
- C. Sắp xếp dữ liệu theo ngày tăng dần
- D. Tính số ngày giữa các dòng

**💡 Đáp án: B** — Không có `parse_dates`, pandas đọc ngày tháng như chuỗi text bình thường và không thể tính toán thời gian.

---

**Câu 5.** `validate="many_to_one"` trong `.merge()` kiểm tra điều gì?

- A. Nhiều dòng bên trái có thể khớp với 1 dòng bên phải (bình thường)
- B. Cả hai bảng đều phải có key duy nhất
- C. Kết quả merge không được có NaN
- D. Số cột của hai bảng phải bằng nhau

**💡 Đáp án: A** — Ví dụ: 1 khách hàng có nhiều đơn hàng → nhiều dòng trong `orders` khớp với 1 dòng trong `customers`. Nếu `customers` có customer_id trùng lặp, merge sẽ báo lỗi ngay.

### 🟡 Nhóm 2 — Data Cleaning & KPI (Câu 6–10)

---

**Câu 6.** Cột `analysis_revenue` của đơn hàng có `status = "cancelled"` bằng bao nhiêu?

- A. Bằng `net_revenue` gốc của đơn
- B. -1 (dấu trừ thể hiện hoàn tiền)
- C. NaN
- D. 0

**💡 Đáp án: D** — Đơn bị huỷ không đóng góp doanh thu thực tế. Đặt về `0` thay vì xoá dòng vì ta vẫn cần dữ liệu này để phân tích tỷ lệ huỷ đơn.

---

**Câu 7.** AOV là chỉ số gì?

- A. Average Output Value
- B. Average Order Value — giá trị trung bình mỗi đơn hàng
- C. Annual Order Volume
- D. Adjusted Order Value

**💡 Đáp án: B** — AOV = Tổng doanh thu ÷ Số đơn hàng. Chỉ số này cho biết mỗi đơn trung bình đem lại bao nhiêu tiền.

---

**Câu 8.** Hàm `.pct_change()` tính gì?

- A. Tổng tích lũy từ đầu đến hiện tại
- B. Phần trăm thay đổi so với dòng liền trước
- C. Trung bình động 3 tháng
- D. Tỷ lệ so với tổng cột

**💡 Đáp án: B** — Ví dụ: tháng 1 revenue = 100, tháng 2 = 120 → `.pct_change()` = +20%.

---

**Câu 9.** `.nunique()` trả về giá trị nào?

- A. Tổng số dòng của cột
- B. Số giá trị bị null
- C. Số giá trị duy nhất (không lặp lại)
- D. Số cột trong DataFrame

**💡 Đáp án: C** — `.nunique()` đếm có bao nhiêu giá trị khác nhau. Dùng để đếm số khách hàng, số sản phẩm riêng biệt, ...

---

**Câu 10.** `observed=True` trong `.groupby()` có tác dụng gì?

- A. Hiển thị tất cả nhóm kể cả nhóm không có dữ liệu
- B. Chỉ tính những nhóm có dữ liệu thực tế, bỏ qua nhóm trống
- C. Sắp xếp kết quả theo thứ tự tăng dần
- D. Loại bỏ NaN trước khi group

**💡 Đáp án: B** — Khi cột groupby có kiểu Categorical, pandas mặc định tạo ra tất cả tổ hợp có thể kể cả tổ hợp không có dữ liệu. `observed=True` giúp bỏ qua những nhóm trống đó.

- C. Chuyển dữ liệu từ dạng dài (nhiều dòng) thành bảng ma trận (hàng × cột)
- D. Tính trung bình cộng của một cột

**💡 Đáp án: C** — `pivot_table` biến dữ liệu từng dòng thành bảng 2 chiều. Ví dụ: từ cột `cohort_index` và `first_purchase_month` → thành ma trận retention.

```python
# ═══════════════════════════════════════════════════════════════════════════
# Submission Artifact Checker
# Tự động kiểm tra tất cả file cần nộp
# ═══════════════════════════════════════════════════════════════════════════

from pathlib import Path

# Notebook này nằm tại: Cohort & Retention Analytics Portfolio\
# Project thực tế nằm tại: cohort-retention-analytics\ (cùng cấp với folder này)
PROJECT_DIR = Path.cwd().parent / "cohort-retention-analytics"

print(f"📁 Project folder: {PROJECT_DIR}")
print(f"   Tồn tại: {'✅' if PROJECT_DIR.exists() else '❌ KHÔNG TÌM THẤY'}")
print()

# Danh sách file cần có để nộp bài
REQUIRED_FILES = [
    # Notebooks
    ("notebooks/project_starter.ipynb",            "Notebook phân tích chính"),
    # Tables
    ("outputs/tables/monthly_kpi.csv",             "KPI theo tháng"),
    ("outputs/tables/channel_contribution.csv",    "Revenue theo kênh"),
    ("outputs/tables/cohort_retention_matrix.csv", "Retention matrix"),
    ("outputs/tables/cohort_revenue_matrix.csv",   "Revenue cohort matrix"),
    ("outputs/tables/data_quality_report.csv",     "Báo cáo chất lượng dữ liệu"),
    ("outputs/tables/validation_checks.csv",       "Kết quả kiểm tra dữ liệu"),
    # Charts
    ("outputs/charts/monthly_revenue_trend.png",   "Chart doanh thu theo tháng"),
    ("outputs/charts/revenue_by_channel.png",      "Chart doanh thu theo kênh"),
    ("outputs/charts/cohort_retention_heatmap.png","Chart heatmap retention"),
    # Reports
    ("reports/executive_summary.md",               "Báo cáo tóm tắt kết quả"),
    ("reports/business_brief.md",                  "Business questions"),
    # Root
    ("README.md",                                  "Hướng dẫn chạy project"),
    ("requirements.txt",                           "Danh sách thư viện"),
]

passed = 0
failed = []

print("─" * 65)
print(f"  {'ST':<4}  {'File':<42}  Mô tả")
print("─" * 65)

for rel_path, description in REQUIRED_FILES:
    full_path = PROJECT_DIR / rel_path
    ok = full_path.exists()
    status = "✅" if ok else "❌"
    print(f"  {status}     {rel_path:<42}  {description}")
    if ok:
        passed += 1
    else:
        failed.append(rel_path)

print("─" * 65)
print(f"\n📊 Kết quả: {passed}/{len(REQUIRED_FILES)} file có sẵn")

if passed == len(REQUIRED_FILES):
    print("\n🎉 TẤT CẢ FILE ĐỦ ĐỦ — Bạn đã sẵn sàng nộp bài!")
else:
    print(f"\n⚠️  Còn {len(failed)} file chưa có:")
    for f in failed:
        print(f"   ❌ {f}")
    print("\n💡 Gợi ý: Mở project_starter.ipynb và chạy lại section còn thiếu.")
```
