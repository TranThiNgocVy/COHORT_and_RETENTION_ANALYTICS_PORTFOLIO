### 🔍 Vấn đề cốt lõi: Acquisition vs. Retention

Hãy nghĩ đơn giản như thế này:

```
📈 Doanh thu tháng 3/2026: 500 triệu VNĐ
├── 🆕 Từ khách hàng MỚI (acquisition):  420 triệu  (84%)
└── 🔄 Từ khách hàng CŨ quay lại (retention):  80 triệu  (16%)

💸 Chi phí marketing để kéo khách mới tháng 3/2026: 200 triệu
   → ROI thực tế: (420 - 200) / 200 = 110%  ← Có vẻ ổn?

❓ Nhưng nếu tháng 4/2026 không chi 200 triệu nữa thì doanh thu còn lại bao nhiêu?
   → Chỉ còn ~80 triệu (retention) + organic = rất thấp!
```

> **💡 Insight quan trọng:**  
> Một business **phụ thuộc quá nhiều vào acquisition** rất dễ tổn thương.  
> Business **bền vững** cần tỷ lệ retention cao — khách hàng tự quay lại mà không cần tốn nhiều marketing budget.

### 🧩 Hiểu từng khái niệm quan trọng

**📌 Cohort là gì?**

> Một **cohort** là nhóm khách hàng được gộp lại theo cùng một mốc thời gian — thường là **tháng mua hàng đầu tiên**.

Ví dụ:
```
Cohort Jan-2026:  Tất cả khách hàng mua lần đầu trong tháng 1/2026
Cohort Feb-2026:  Tất cả khách hàng mua lần đầu trong tháng 2/2026
...
```

**📌 Retention Rate là gì?**

> **Retention Rate** = % khách hàng của một cohort **quay lại mua hàng** ở tháng tiếp theo, tháng sau nữa...

Ví dụ đọc cohort matrix — Cohort Jan-2026 (100 khách mua lần đầu trong tháng 1/2026):

```
Cohort Index  │  Tháng thực tế  │  Số khách active   │         Retention Rate
──────────────┼─────────────────┼────────────────────┼────────────────────────────────
   Index 0    │   Jan/2026      │     100 khách      │   100% ← tháng gia nhập (cơ sở)
   Index 1    │   Feb/2026      │      42 khách      │    42% ← 1 tháng sau gia nhập
   Index 2    │   Mar/2026      │      28 khách      │    28% ← 2 tháng sau gia nhập
   Index 3    │   Apr/2026      │      19 khách      │    19% ← 3 tháng sau gia nhập
```

> ⚠️ **Giải thích "Index 0, 1, 2...":**  
> Đây là **cohort index** — số tháng tính từ lần mua đầu tiên của nhóm, KHÔNG phải số thứ tự tháng lịch.  
> - Index 0 = tháng gia nhập cohort (Jan/2026)  
> - Index 1 = 1 tháng sau gia nhập (Feb/2026)  
> - Index 2 = 2 tháng sau gia nhập (Mar/2026) … và tiếp tục

> **Câu hỏi tự kiểm tra:** Nếu Cohort Jan-2026 có retention Index 1 = 42%, còn Cohort Feb-2026 có retention Index 1 = 61% thì tháng nào kéo được khách "chất lượng" hơn? Tại sao?

**📌 Tại sao phân tích Cohort tốt hơn chỉ nhìn tổng doanh thu?**

Hãy xem ví dụ sau:

```
Tháng 3/2026:  Tổng khách mua = 1,000  ← trông có vẻ ổn
Tháng 4/2026:  Tổng khách mua = 1,200  ← tăng 20%!

Nhưng nếu nhìn kỹ hơn:
  Tháng 4/2026:  Khách mới = 1,150   Khách cũ quay lại = 50
  → Retention Rate ≈ 5%  ← RẤT THẤP!

→ Kết luận: tăng trưởng là giả tạo, phụ thuộc hoàn toàn vào acquisition.
```

```
cohort-retention-analytics/          ← Root folder
│
├── data/
│   ├── raw/                          ← Dữ liệu gốc — KHÔNG chỉnh sửa
│   │   ├── customers.csv
│   │   ├── orders.csv
│   │   ├── order_items.csv
│   │   └── products.csv
│   └── processed/                    ← Dữ liệu sau khi làm sạch
│       └── clean_orders.csv
│
├── notebooks/
│   └── project_starter.ipynb         ← Notebook làm việc chính
│
├── outputs/
│   ├── tables/                       ← KPI tables, cohort matrices
│   │   ├── monthly_kpi.csv
│   │   ├── channel_contribution.csv
│   │   ├── cohort_retention_matrix.csv
│   │   ├── cohort_revenue_matrix.csv
│   │   ├── data_quality_report.csv
│   │   └── validation_checks.csv
│   └── charts/                       ← Biểu đồ xuất ra PNG
│       ├── monthly_revenue_trend.png
│       ├── revenue_by_channel.png
│       └── cohort_retention_heatmap.png
│
├── reports/
│   ├── business_brief.md             ← Business context & câu hỏi phân tích
│   └── executive_summary.md          ← Báo cáo 1 trang cho stakeholder
│
├── src/
│   └── generate_ecommerce_dataset.py ← Script sinh dữ liệu tổng hợp
│
├── README.md                         ← Mô tả project + cách chạy
└── requirements.txt                  ← Thư viện cần cài
```

> 💡 **Tại sao `data/raw/` tách riêng với `data/processed/`?**  
> Nguyên tắc vàng: **không bao giờ ghi đè dữ liệu gốc**. Nếu code làm sạch có lỗi, bạn luôn có thể quay về `raw/` để bắt đầu lại mà không mất dữ liệu gốc.

### 📝 Câu 2 — Phân tích tình huống

Giả sử báo cáo tháng 6 cho thấy:
- Tổng doanh thu tháng 6: **800 triệu**
- Doanh thu từ khách mới: **750 triệu**
- Doanh thu từ khách cũ quay lại: **50 triệu**
- Chi phí marketing tháng 6: **300 triệu**

**Câu hỏi:** Đây là acquisition growth hay retention growth? Business này có vấn đề gì? Nếu tháng 7 cắt toàn bộ chi phí marketing, điều gì sẽ xảy ra?

---

**💡 Gợi ý đáp án:**  
- Đây là **acquisition growth** — 93.75% doanh thu đến từ khách mới.  
- Vấn đề: retention cực thấp (~6.25%), phụ thuộc hoàn toàn vào marketing spend.  
- Nếu cắt marketing tháng 7: doanh thu sẽ giảm đột ngột về ~50 triệu hoặc thấp hơn.
